```python
import pickle
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier

from sklearn.metrics import classification_report, accuracy_score
```

```python
df = pd.read_csv("reshuffled_final_final_final.csv")
df.head()
```

```python
X=df[['Medical_Issue1','Medical_Issue2','Medical_Issue3','Caretaker_Expertise']]
X
```

```python
y=df['Recommended']
y
```

```python
df_encoded = pd.get_dummies(df[['Medical_Issue1','Medical_Issue2','Medical_Issue3','Caretaker_Expertise']],dtype='int')
```

```python
df_encoded
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_encoded, y, test_size=0.2, random_state=101)
```

```python
models = {
    "Logistic Regression": LogisticRegression(),
    "Random Forest": RandomForestClassifier()
}

# Train models and evaluate performance
for model_name, model in models.items():
    print(f"\nTraining {model_name}...")

    # Train model
    model.fit(X_train, y_train)

    # Predict and evaluate
    y_pred = model.predict(X_test)
    
    # Print accuracy and classification report
    print(f"{model_name} Accuracy: {accuracy_score(y_test, y_pred):.4f}")
    print(classification_report(y_test, y_pred))
    
# Optionally, perform hyperparameter tuning with GridSearchCV (Example for Random Forest)
param_grid_rf = {
    'n_estimators': [400,410,420],
    'max_depth': [40,50,30],
    'min_samples_split': [4,3,5]
}
grid_search_rf = GridSearchCV(RandomForestClassifier(), param_grid_rf, cv=3, scoring='accuracy')
grid_search_rf.fit(X_train, y_train)
print(f"Best Random Forest Params: {grid_search_rf.best_params_}")

# Train the optimized Random Forest model
best_rf = grid_search_rf.best_estimator_
best_rf.fit(X_train, y_train)
y_pred_rf = best_rf.predict(X_test)
print(f"Optimized Random Forest Accuracy: {accuracy_score(y_test, y_pred_rf):.4f}")
```

```python
with open('model.pkl', 'wb') as model_file:
    pickle.dump(best_rf, model_file)

print("Model saved as caretaker_recommendation_model.pkl")
```
